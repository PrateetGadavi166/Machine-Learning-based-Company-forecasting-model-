```python
#splitting Data into Training and Testing

# 70% data: training set of values of 'close' ranging from 0 to 70% of data
data_training = pd.DataFrame(df['Close'][0:int(len(df)*0.70)])

# 30% data: testing set of values of 'close' ranging from 70% to whole data
data_testing = pd.DataFrame(df['Close'][int(len(df)*0.70): int(len(df))])

print(data_training.shape)
print(data_testing.shape)
```

```python
data_training.head()
```

```python
data_testing.head()
```

```python
#scaling down the data 0 and 1 using 'MinMaxScaler'
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
```

```python
#converting to array of both data_training and data_testing which has been just now fitted
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
# 2 empty list of trains are created on x and y
x_train = []
y_train = []

#insert values in the list
#this 100 is to predict next value
for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100 : i])
    y_train.append(data_training_array[i, 0])
    
#convert into both x and y train to numpy arrays to provide data to LSTM
x_train , y_train = np.array(x_train), np.array(y_train)
```

```python
#ML MODEL
```

```python
#!pip install keras
```

```python
#!pip install tensorflow
```

```python
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
#model is created
model = Sequential()

#layers are added, units ,activation function, x_train.shape:(___, 100, 1) 100 because 100 columns/ steps first 100 values have become columns, currently working with 1 column i.e 'Close' so 1.
model.add(LSTM(units = 50, activation = 'relu' , return_sequences = True,
               input_shape = (x_train.shape[1], 1)))
model.add(Dropout(0.2))

#layer1
model.add(LSTM(units = 60, activation = 'relu' , return_sequences = True))
model.add(Dropout(0.3))

#layer2
model.add(LSTM(units = 80, activation = 'relu' , return_sequences = True))
model.add(Dropout(0.4))

#layer3
model.add(LSTM(units = 120, activation = 'relu' ))
model.add(Dropout(0.5))

#to connect all the layers
model.add(Dense(units = 1))
```

```python
#compiled
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
model.fit(x_train , y_train, epochs = 50)
```

```python
#save
#model.save('keras_model.h5')
# Save the model in the recommended Keras format
model.save('my_model.keras')
```
